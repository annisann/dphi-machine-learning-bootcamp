### Separate the Input and Target Features of the data

```python
x = df.drop(columns=['deathrate', 'location', 'med_income_decile', 'edu_college_18to24'])
y = df.deathrate
```

Cuz we want to predict the death rate of cancer, so for the x variabel, we included all columns EXCEPT deathrate.

### Split the data into Train and Test Sets

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=42)
size = pd.DataFrame({'x':[x_train.shape, y_train.shape],'y':[x_test.shape, y_test.shape]},
                    index=['Train Data', 'Test Data'])
size
```

## Task 2

### Build a Linear Regression Model

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(x_train, y_train)

lr_pred = lin_reg.predict(x_test)
lr_predtrain = lin_reg.predict(x_train)
pd.DataFrame(lr_pred)
```

### Evaluate your model with Mean Squared Error (MSE)

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

print('MSE\t', mean_squared_error(y_test, lr_pred))
print('MAE\t', mean_absolute_error(y_test, lr_pred))
print('RMSE\t', np.sqrt(mean_squared_error(y_test, lr_pred)))
```

```python
print('MSE\t', mean_squared_error(y_train, lr_predtrain))
print('MAE\t', mean_absolute_error(y_train, lr_predtrain))
print('RMSE\t', np.sqrt(mean_squared_error(y_train, lr_predtrain)))
```

### Build Random Forest Model

```python
from sklearn.ensemble import RandomForestRegressor

rf_model1 = RandomForestRegressor(random_state=1)
rf_model2 = RandomForestRegressor(random_state=1, max_depth=5)
rf_model3 = RandomForestRegressor(random_state=2)
```

```python
rf_model1.fit(x_train, y_train)
rf_model2.fit(x_train, y_train)
rf_model3.fit(x_train, y_train)
```

```python
rf_model1.fit(x_train, y_train)
```

```python
m1_train = rf_model1.predict(x_train)
m1_test = rf_model1.predict(x_test)

m2_train = rf_model2.predict(x_train)
m2_test = rf_model2.predict(x_test)

m3_train = rf_model3.predict(x_train)
m3_test = rf_model3.predict(x_test)
```

### Evaluate your model with Mean Squared Error (MSE)

```python
print('Random Forest Regressor - Model 1')
print('MSE Train\t', mean_squared_error(y_train, m1_train))
print('MSE Test\t', mean_squared_error(y_test, m1_test))
print('RMSE Train\t', np.sqrt(mean_squared_error(y_train, m1_train)))
print('RMSE Test\t', np.sqrt(mean_squared_error(y_test, m1_test)))

print('\nRandom Forest Regressor - Model 2')
print('MSE Train\t', mean_squared_error(y_train, m2_train))
print('MSE Test\t', mean_squared_error(y_test, m2_test))
print('RMSE Train\t', np.sqrt(mean_squared_error(y_train, m2_train)))
print('RMSE Test\t', np.sqrt(mean_squared_error(y_test, m2_test)))

print('\nRandom Forest Regressor - Model 3')
print('MSE Train\t', mean_squared_error(y_train, m3_train))
print('MSE Test\t', mean_squared_error(y_test, m3_test))
print('RMSE Train\t', np.sqrt(mean_squared_error(y_train, m3_train)))
print('RMSE Test\t', np.sqrt(mean_squared_error(y_test, m3_test)))
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor

dtree_model1 = DecisionTreeRegressor(random_state = 1, max_depth = 5)
dtree_model2 = DecisionTreeRegressor(random_state = 1, max_depth = 10)
dtree_model3 = DecisionTreeRegressor(random_state = 1, max_depth = 15)
```

```python
dtree_model1.fit(x_train, y_train)
dtree_model2.fit(x_train, y_train)
dtree_model3.fit(x_train, y_train)
```

```python
print('Decision Tree Regressor - Model 1')
print('RMSE Train\t', np.sqrt(mean_squared_error(y_train, dt1_train)))
print('RMSE Test\t', np.sqrt(mean_squared_error(y_test, dt1_test)))

print('\nDecision Tree Regressor - Model 2')
print('RMSE Train\t', np.sqrt(mean_squared_error(y_train, dt2_train)))
print('RMSE Test\t', np.sqrt(mean_squared_error(y_test, dt2_test)))

print('\nDecision Tree Regressor - Model 3')
print('RMSE Train\t', np.sqrt(mean_squared_error(y_train, dt3_train)))
print('RMSE Test\t', np.sqrt(mean_squared_error(y_test, dt3_test)))
```

## Task 3

### Perform Hyperparameter Tunning on Random Forest Model

```python
from sklearn.model_selection import GridSearchCV

parameters = {
    'max_depth': [10, 20],
    'min_samples_leaf': [3, 4],
    'min_samples_split': [4, 6],
    'n_estimators': [100, 200],
}

grid = GridSearchCV(estimator=rf_model1, param_grid=parameters, cv = 3, n_jobs = -1, verbose = 2)
grid.fit(x_train, y_train)

grid.best_estimator_
```

```python
rf_modelcv = grid.best_estimator_
rf_modelcv.fit(x_train, y_train)

mcv_train = rf_modelcv.predict(x_train)
mcv_test = rf_modelcv.predict(x_test)
```

### Evaluate your model with Mean Squared Error (MSE)

```python
print('Random Forest with GridSearchCV - Train Data')
print('MSE\t', mean_squared_error(y_train, mcv_train))
print('MAE\t', mean_absolute_error(y_train, mcv_train))
print('RMSE\t', np.sqrt(mean_squared_error(y_train, mcv_train)))

print('\nRandom Forest with GridSearchCV - Test Data')
print('MSE\t', mean_squared_error(y_test, mcv_test))
print('MAE\t', mean_absolute_error(y_test, mcv_test))
print('RMSE\t', np.sqrt(mean_squared_error(y_test, mcv_test)))
```

```python
from boruta import BorutaPy

boruta_select = BorutaPy(rf_model1, n_estimators='auto', verbose=2, random_state=1 )
boruta_select.fit(np.array(x_train), np.array(y_train))
```

```python
print("Selected Features\t", boruta_select.support_)
print("Ranking\t",boruta_select.ranking_)
print("No. of significant features\t", boruta_select.n_features_)
```

```python
# Let's visualise it better in the form of a table
selected_features = pd.DataFrame({'Feature':list(x_train.columns),
                                      'Ranking':boruta_select.ranking_})
selected_features.sort_values(by='Ranking')
```

```python
x_important_train = boruta_select.transform(np.array(x_train))
x_important_test = boruta_select.transform(np.array(x_test))

# Create a new random forest classifier for the most important features
rf_important = RandomForestRegressor(random_state=1, n_estimators=1000, n_jobs = -1)
```

```python
# Train the new classifier on the new dataset containing the most important features
rf_important.fit(x_important_train, y_train)

rf_important_pred_train = rf_important.predict(x_important_train)

print('Random Forest After Feature Selection (Boruta) - Train Data')
print('MSE\t', mean_squared_error(y_train, rf_important_pred_train))
```

```python
rf_important.fit(x_important_test, y_test)

rf_important_pred_test = rf_important.predict(x_important_test)

print('Random Forest After Feature Selection (Boruta) - Test Data')
print('MSE\t', mean_squared_error(y_test, rf_important_pred_test))
```
